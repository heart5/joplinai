# 魔法路径

## 引入库

```python
import sys
```

## context类

```python
class context:
    def __enter__(self):
        sys.path.extend(["."])

    def __exit__(self, *args):
        pass
```

## 函数集

## 主函数main()

```python
if __name__ == "__main__":
    if "__file__" in list(locals()):
        print(f"运行文件\t{__file__}")

    for pp in sys.path:
        print(pp)

    print("It's man, very good.")
    print("Done.完毕。")
```
